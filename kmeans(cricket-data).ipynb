# OUTLIERS HANDALING

```python
def whiskers(col):
    q1=np.quantile(col,0.25)
    q3=np.quantile(col,0.75)
    iqr=q3-q1
    uw=q3+1.5*iqr
    lw=q1-1.5*iqr
    return uw,lw
```

```python
print(whiskers(df["Runs.x"]))
run_out=df[df["Runs.x"]>505.0].index
df.loc[run_out,"Runs.x"]=505.0
```

# SCALING

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df=ss.fit_transform(df)
```

# CLUSTERING

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=1)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

# FIND N_CLUSTER USING ELBOW METHOD

```python
plt.plot(range(1,11),wcss,"o--")
plt.grid()
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,random_state=1)
ylabel=kmeans.fit_predict(df)
```

# CLASIFICATION

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
def algorithm(algo):
    algo.fit(xtrain,ytrain)
    ypred=algo.predict(xtest)
    
    train=algo.score(xtrain,ytrain)
    test=algo.score(xtest,ytest)
    print(f"TrainingAccuracy: {train}\n Testin Accuracy: {test}\n\n")
    print(classification_report(ytest,ypred))
    return algo
from sklearn.metrics import classification_report 
```
